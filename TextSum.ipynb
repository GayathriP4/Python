## **1.Word/Term frequency**

```
#importing nltk & downloading all the packages in it
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```
#stopwords for the built-in set of stopwords
#Porter Stemmer for stemming
#pandas for dataframe -  to visualize the data
#word_tokenize & sent_tokenize are built-in modules for word tokenization & sentence tokenization

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
```

```
def word_freq(words):
  d = {}
  sw = set(stopwords.words('english'))
  ps = PorterStemmer()
  nw = 0

  for word in words:
    w = word.lower()
    if w not in sw:
      #stopword removed
      nw+=1
      if w.isalpha():
        w = ps.stem(w)
      
      if w not in d:
      #not there in dictionary
        d[w] = 1 #count set to 1
      else:
      #already exists in dictionary
        d[w] = d[w] + 1 #increment count by 1
  
  return d,nw 
  #d -dictionary that holds the word frequency
  #nw - total no. of words
```

```
wt = word_tokenize(text)
d,nw = word_freq(wt)
d
```

```
df = pd.DataFrame(list(d.items()),columns = ['word','freq']) 
df.head(10)
```

```
df = pd.DataFrame(list(d.items()),columns = ['word','freq']) 
df['norm-freq'] = [d[i]/nw for i in d]
d = {i:d[i]/nw for i in d}
df
```

## 2.Sentence tokenize

```
st = sent_tokenize(text)
st
```

## 3.Score the sentences

```
ss = {}
ns = len(st)
ps = PorterStemmer()
for sent in st:
  #considering each sentence
  #words in that sentence
  w1 = word_tokenize(sent)
  c = 0
  for word in w1:
    if word.isalpha():
      word_stem = ps.stem(word.lower())
    else:
      word_stem = word
    if word_stem in d:
      c = c + d[word_stem]
  #ss[sent] = [c,len(w1)]
  ss[sent] = {}
  ss[sent]['score'] = c
  ss[sent]['total'] = len(w1)
ss  
```

```
#sdf = pd.DataFrame(list(ss.items()),columns = ['sentence','score']) 
sdf = pd.DataFrame()
sdf['sentence'] = [i for i in ss] 
sdf['score'] =  [ ss[i]['score'] for i in ss]
sdf['total'] = [ ss[i]['total'] for i in ss]
sdf
```

## 4. AVG Score

```
fs = {}
total_score = 0
l = len(ss)
for sent in ss:
  score = ss[sent]['score']
  total = ss[sent]['total']
  x = score * total  
  fs[sent] = x
  total_score+=x
avg_score = total_score/l
fs
```

```
df1 = pd.DataFrame(list(fs.items()),columns = ['sentence','score']) 
df1
```
