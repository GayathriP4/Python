```
#all the required imports
#only once
import os
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
from nltk.stem import PorterStemmer as PS
from nltk.stem import WordNetLemmatizer as wnl
import pandas as pd
```

```
def pp(n):
    f = open(n,'r',encoding='latin-1')
    i = f.read()
    #print(n)
    f.close()
    #print(n)

    # removing punctuation
    t = "".join(c for c in i if c not in string.punctuation)
    
    # converting to tokens
    t = word_tokenize(t)
    
    # removing numbers and converting to lowercase
    t = [ x.lower() for x in t if x.isalpha() ]
    
    #removing stopwords and words with length 1 or greater than 14
    sw = set(stopwords.words('english'))
    tex = [ x for x in t if x not in sw and len(x)>1 and len(x)<15 ]
    
    #lemmatization
    l = wnl()
    text = [ l.lemmatize(x) for x in tex ]
   
   #stemming
    p = PS()
    tex = [ p.stem(x) for x in text ]
  
    #tex.sort()
    d = {}
    sum = 0
    for x in tex:
      #if d.has_key(x):
      if x in d:
        d[x] = d[x] + 1
        sum+=1
      else:
        d[x] = 1
        sum+=1
    return (d,sum)
```

```
def lfld(path,fp):
    if os.path.isdir(path):
        l = os.listdir(path)
        for x in l:
            tp = path + '/' + x
            lfld(tp,fp)
    elif os.path.isfile(path):
        fp.append(path)
        return
    else: pass
```

```
#path = '/content/gdrive/My Drive/np/train'
path = '/content/gdrive/My Drive/np/train/sci.crypt'
fp = []

print("Fetching Files ....")
lfld(path,fp)
print("\nFetched Files ....")
```

```
sd = {}
tfd = {}

print("\nPreprocessing begins ....")
for x in fp:
    ( tfd[x] , sd[x] ) = pp(x)
print("\nPreprocessing done ....")
print('tfd',tfd) #contains how many times a word appears per doc
print('sd',sd) #contains total words in a doc
```

```
print("\nCalculating TF ....")
TF = {}
for d in tfd:
    TF[d] = {}
    for w in tfd[d]:
        TF[d][w] = float(tfd[d][w]) / sd[d]
TFD = { 'doc' : [] , 'tf' : [] }
for d in TF:
    TFD['doc'].append(d)
    TFD['tf'].append(TF[d])

print("\nTF as a DataFrame")
TF = pd.DataFrame(TFD)
print('TF',TF) #
print('TFD',TFD) #
```

```
print("\nCalculating DF ....")
dfd = {}
for ok in tfd:
    for ik in tfd[ok]:
      if ik in dfd:
      #if dfd.has_key(ik):
        dfd[ik].append(ok)
      else:
        dfd[ik] = [ok]
DFD = { 'word' : [] , 'doc' : [] }
for key in sorted(dfd):
    #print(key,len(dfd[key]))
    DFD['word'].append(key)
    DFD['doc'].append(dfd[key])

print("\nDF as a DataFrame ")
DF = pd.DataFrame(DFD)
#print('DF',DF)
#print('DFD',DFD)
```

```
def lfld(path,fp):
    if os.path.isdir(path):
        l = os.listdir(path)
        for x in l:
            tp = path + '/' + x
            lfld(tp,fp)
    elif os.path.isfile(path):
        fp.append(path)
        return
    else: pass
```

```
def pp(n):
  f = open(n,'r',encoding='latin-1')
  i = f.read()
  f.close()
  # removing punctuation
  t = "".join(c for c in i if c not in string.punctuation)
    
  # converting to tokens
  t = word_tokenize(t)
    
  # removing numbers and converting to lowercase
  t = [ x.lower() for x in t if x.isalpha() ]
    
  #removing stopwords and words with length 1 or greater than 14
  sw = set(stopwords.words('english'))
  tex = [ x for x in t if x not in sw and len(x)>1 and len(x)<15 ]
    
  #lemmatization
  l = wnl()
  text = [ l.lemmatize(x) for x in tex ]
   
  #stemming
  p = PS()
  tex = [ p.stem(x) for x in text ]
  t = [ " ".join(c for c in tex) ]
  #print(t)
  return t
```

```
def tf_idf(fp):
  vectorizer = TfidfVectorizer()
  response = vectorizer.fit_transform(fp)
  fname = vectorizer.get_feature_names()
  vc = response[0]
  df = pd.DataFrame(vc.T.todense(),fname, columns=["tfidf"])
  df = df.sort_values(by=["tfidf"],ascending=False)
  #print(fp)
  #print(df.head(20))
  return df
```

```
path = '/content/gdrive/My Drive/np/train/sci.crypt'
fp = []
print("Fetching Files...")
lfld(path,fp)
print("Files fetched from ",path,end='\n\n')
tfidf = {}
for f in fp:
  s = pp(f)
  tfidf[f] = tf_idf(s)
print("Preprocessing done and tf-idf calculated.")
```
